With this dataset, I'd like to explore whether the ingredients used in recipes originating from certain countries would be similar based on geographical proximity. Intuitively, I believe they will. As one could imagine, historically, availability of certain ingredients would be geographically based. In modern times, the techniques for the transportation of food has become increasingly advanced and this would lead to much more crossover between geographical regions. This trend likely decreases the degree to which regions' recipes will be separated, but I would still guess the geographical differences will be apparent.

---------------------------

I begin my analysis by parsing the data into a matrix with the rows being the available countries and the columns being available ingredients. I then divide these raw frequencies by the number of recipes available from that country to account for variability in the number of recipes attributed to different countries. 

```python
import numpy as np
import csv
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics
import scipy.cluster.hierarchy as hr
from scipy.spatial.distance import pdist, squareform
import matplotlib
matplotlib.use('cairo')
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# set to contain all unique ingredients 
allIng = set()
# set to contain all unique countries
countries = set()
files = ['allr', 'epic', 'menu']

# find all possible countries and ingredients
for f in files:
    with open("./scirep-cuisines-detail/%s_recipes.txt" % f, 'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter='\t')
        for row in reader:
            for i in range(1,len(row)):
                allIng.add(row[i])
                countries.add(row[0].lower())
            
allIng = list(allIng)
zeros = np.zeros((len(countries), len(allIng)))
data = pd.DataFrame(data=zeros, index=countries, columns=allIng)
# dict to store counts of recipes originating in each country
countryCnt = dict((i,0) for i in countries)

# I'm not happy that I iterate through these files again, but it works. 
# will revisit if time allows
for f in files:
    with open("./scirep-cuisines-detail/%s_recipes.txt" % f, 'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter='\t')
        for row in reader:
            for i in range(1,len(row)):
                c = row[0].lower() #country name to lowercase to avoid confusion
                curr = data.get_value(c, row[i])
                
                countryCnt[c] = countryCnt[c]+1

                if type(curr) == float:
                    data.set_value(c, row[i], 1)
                else: 
                    data.set_value(c, row[i], curr+1)

cnt = pd.Series(countryCnt)
```

I noticed some repeating countries with slightly different names, like China versus Chinese, or Mexico versus Mexican. I combine these data points here.

```python
# Combine duplicate entries (China v Chinese, etc)
# This code will only execute once. If it throws an error, 
# rerun the previous code cell or restart the kernel

dupes = [['china', 'chinese'],
         ['france', 'french'],
         ['germany', 'german'],
         ['india', 'indian'],
         ['italy', 'italian'],
         ['japan', 'japanese'],
         ['korea', 'korean'],
         ['mexico', 'mexican'],
         ['scandinavia', 'scandinavian'],
         ['thailand', 'thai'],
         ['vietnam', 'vietnamese']]

for p1, p2 in dupes:
    data.loc[p1] = (data.loc[p1] + data.loc[p2])
    data = data[data.index != p2]
    cnt[p1] = (cnt[p1] + cnt[p2])
    cnt = cnt[cnt.index != p2]
```

Finally, I scale the data and print it here in the representation I will use.

```python
# scaling the data by the number of recipes in the dataset
data_scaled = data.div(cnt, axis='rows')
print data_scaled
```

Now that our data is in usable form, I've hierarchically clustered it and plotted both a dendogram and a heatmap. The dendogram shows some interesting results, with Bangladesh's recipes being very dissimilar from the rest and the similarly isolated Austria and Netherlands cluster. I've highlighted the dendogram to show what I feel to be the most appropriate clustering.

```python
# compute the clustering
distances = metrics.euclidean_distances(data_scaled)
clust = hr.linkage(distances, method='complete', metric='euclidean')
```

```python
fig = plt.figure(figsize=(10,10))
T = hr.dendrogram(clust,color_threshold=0.37, leaf_font_size=10)

# reorder country name list to display properly on axis
x = list(data_scaled.index)
x.sort()
countOrd = [x[i] for i in T['leaves']]

plt.xlabel('Country of Origin')
plt.xticks(plt.xticks()[0], countOrd)
fig.show()
```

Below is the heatmap, where I've plotted countries against countries with the color of the cell representing the distance between their recipe-ingredient frequencies. While both axes labels are the same (as evidenced by the 0 value diagonal) I've put the cluster labels along the x-axis to help visualize the connection between this graphic and the dendogram above.

```python
ck = hr.fcluster(hr.linkage(distances, method='complete'), 0.37,'distance')
new = []
for i in range(len(ck)):
    new.append([ck[i], list(data_scaled.index)[i]])

new.sort()
new = np.array(new)

df = pd.DataFrame(distances)
df.index = ck
df.columns = ck

df = df.sort(ascending=True, axis=1)
df = df.sort(ascending=True, axis=0)
sns.set_context("notebook", font_scale=0.8)
sns.heatmap(df, yticklabels=new[:,1]) #, xticklabels=new[:,1])
```

```python
cDict = {}
for x in range(len(new)):
    if new[x][0] not in cDict:
        cDict[new[x][0]] = [new[x][1]]
    else:
        cDict[new[x][0]].append(new[x][1])
```

```python
regions = {}
with open('./scirep-cuisines-detail/map.txt', 'rb') as csvfile:
    reader = csv.reader(csvfile, delimiter='\t')
    for row in reader:
        if len(row) > 1 and row[1] != '':
            if row[1] not in regions:
                regions[row[1]] = [row[0].lower()]
            else:
                regions[row[1]].append(row[0].lower())

dupes = [['china', 'chinese'],
         ['france', 'french'],
         ['germany', 'german'],
         ['india', 'indian'],
         ['italy', 'italian'],
         ['japan', 'japanese'],
         ['korea', 'korean'],
         ['mexico', 'mexican'],
         ['scandinavia', 'scandinavian'],
         ['thailand', 'thai'],
         ['vietnam', 'vietnamese']]
dupes = np.array(dupes)

print "Defined Regions: "

for i in regions.keys():
    for val in regions[i]:
        if val in dupes[:,1]:
            regions[i].remove(val)
    print i, regions[i]
    
print ""
print "As clustered:"
for key in cDict.keys():
    print key, cDict[key]
```

Here I've printed the regions as defined in our datafile, for reference, and the output of our clustering.

Looking at our first cluster, we see North American countries and Western European countries, as well as Eastern European and one Middle Eastern. This is all very logical, given the colonization of North America (especially the more northern countries, which are clustered here) by Western European powers. Another influencing factor is likely immigration of Eastern Europeans and Jewish people to North American countries. 

Cluster 3, containing no real surprises, is comprised of two Western European countries. Clusters 4 and 5 are also fairly predictable and consistent with my hypothesis, consisting of largely European and Asian countries, respectively. 

Cluster 6 is an interesting combination of Middle Eastern and African countries. I really don't know what to attribute this to, except perhaps the after effects of the Ottoman Empire or similar climates. This would be interesting to explore further with someone more knowledgeable of the area's history. 

Cluster 7 is Bangladesh all on its lonesome. A quick look up at our dendogram shows that were we to raise the cutoff to a point where Bangladesh would join another cluster, it would join 5 and 6, the largely Asian and Middle Eastern clusters. This is somewhat consistent with the geography, where Bangladesh is at an interesting point close to the intersection of Asia and the Middle East. Interestingly, the region of South Asia was entirely split up, with not a single member of that region ending up in the same cluster, perhaps due to the huge geographical area that those three countries span.

Cluster 2 is where it gets interesting. North American, African, Latin American, Southern European, Southeast Asian, and Middle Eastern, all in the same, largest group. Even taking a look at our dendogram, we have to go back several cluster joinings before seeing a geographical split. Much of these similarities are likely due to the original colonization of the Americas and the slave labor used by countries such as Portugal, Spain, and other Mediterranean powers. The majority of these slaves were sent to Latin American and southern areas of North America, where their labor was used to produce valuable cash crops of the time (http://en.wikipedia.org/wiki/Atlantic_slave_trade). These interactions over many years likely led to the similarities in food we see here today. Strangely, we also see India, Malaysia and Middle Eastern recipes. India and Malaysia have also experienced colonial forces throughout their histories which could have led to some of these similarities.

Overall, this is an interesting study in the development of diets and food preparation around the world. An interesting next step to take would be to look at it through the lens of agriculture and perhaps an even longer historical scope.
